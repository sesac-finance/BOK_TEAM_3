```python
import pandas as pd
import re
import csv
from ekonlpy.tag import Mecab
from ekonlpy.sentiment import MPCK
mpck = MPCK()

#2011~2020까지 10회 반복(파일 당 1년)\n

article=pd.read_csv("refinal3.csv", header=0, encoding='utf-8')
article.insert(2,'tokens_ngrams',None)
```

```python
article
```

```python
for i in range(150000,len(article.data)): # len(article)\
    print(i)
    try:
        piece = str(article.iloc[i,0])
        # piece = tidy_sentences(piece)
        new_content=[]

        tokens = mpck.tokenize(piece)
        ngrams = mpck.ngramize(tokens)
        new_content+=tokens
        new_content+=ngrams
        article['tokens_ngrams'][i]=new_content
        
    except:
        
        article['tokens_ngrams'][i]=None

article.to_csv('naver_ngram_2022.csv', encoding='utf-8', index=False)
```
